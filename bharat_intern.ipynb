# Preprocessing
### 1. Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_train['Segment'] = le.fit_transform(df_train['Segment'])
df_train['City'] = le.fit_transform(df_train['City'])
df_train['State'] = le.fit_transform(df_train['State'])
df_train['Region'] = le.fit_transform(df_train['Region'])
df_train['Category'] = le.fit_transform(df_train['Category'])
df_train['Sub-Category'] = le.fit_transform(df_train['Sub-Category'])
df_train['Product Name'] = le.fit_transform(df_train['Product Name'])
df_train['Postal Code'] = le.fit_transform(df_train['Postal Code'])
df_train['Order Date'] = le.fit_transform(df_train['Order Date'])
df_train['Product ID'] = le.fit_transform(df_train['Product ID'])
```

```python
df_train
```

# Splitting data into train and test

```python
X = df_train.drop('Sales',axis=1)
Y = df_train['Sales']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=101, test_size=0.2)
```

```python
X.describe()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
```

```python
X_train_std= sc.fit_transform(X_train)
X_test_std= sc.transform(X_test)
```

```python
X_train_std
```

# Model

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
lr.fit(X_train_std,Y_train)
```

```python
Y_pred_lr=lr.predict(X_test_std)
Y_pred_lr
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(Y_test,Y_pred_lr))
print(mean_absolute_error(Y_test,Y_pred_lr))
print(np.sqrt(mean_squared_error(Y_test,Y_pred_lr)))
```

```python
# Very low correlation (r2_score is less than 0.2)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(n_estimators=1000)
```

```python
rf.fit(X_train_std,Y_train)
```

```python
Y_pred_rf= rf.predict(X_test_std)
Y_pred_rf
```

```python
print(r2_score(Y_test,Y_pred_rf))
print(mean_absolute_error(Y_test,Y_pred_rf))
print(np.sqrt(mean_squared_error(Y_test,Y_pred_rf)))
```

```python
#Random Forest performed better than linear regression
```
